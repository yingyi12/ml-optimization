```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
x = np.random.rand(1000)   #interval (0,1)
noise = 0.5*np.random.randn(1000) # gaussian
y = x+noise

plt.scatter(x,y)

a = np.dot(x,y)/np.dot(x,x)
g = a*x
plt.plot(x,g,c=colors[1])
print(a)
plt.show()
```

```python
x = np.linspace(0,1 ,1000)
z = 30 * (x - 0.25)**2 * (x - 0.75)**2 + 0.1 * np.random.randn(1000)

plt.scatter(x, z, c=colors[0])


z1 = np.polyfit(x, z, 4) 
y1 = np.polyval(z1,x)
plt.plot(x,y1,c=colors[1])


```
